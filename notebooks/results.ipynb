```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ])
ypoints = np.array([76.13, 66.85, 53.82, 23.62, 5.32, 0.86,  0.31])

plt.xlabel("Channel prune amount (%)")
plt.ylabel("Top-1 accuracy (%)")
plt.title("Validation accuracy on the resnet-50 model, fully-connected and convolutional layers pruned.")
plt.grid(axis = 'y')
plt.plot(xpoints, ypoints, marker = 'o', label='Linear + Conv2d layers pruned', color='blue')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

linear_xpoints = np.array([0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ])
linear_ypoints = np.array([76.13, 66.85, 53.82, 23.62, 5.32, 0.86,  0.31])

xpoints = np.array([0, 0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ])
ypoints = np.array([76.13, 69.92, 59.05, 26.96, 6.39, 0.98,  0.34])

plt.xlabel("Channel prune amount (%)")
plt.ylabel("Top-1 accuracy (%)")
plt.title("Validation accuracy on the resnet-50 model.")
plt.grid(axis = 'y')


plt.plot(linear_xpoints, linear_ypoints, marker = 'o', label='Linear + Conv2d layers pruned', color='blue')
plt.plot(xpoints, ypoints, marker = 's', label='Conv2d layers only pruned', color='green')
plt.legend()
plt.show()

```
